# Load libraries and packages

```python
import random
import pandas as pd
```

# Define simulation

```python
# Function to (quickly) return all possible partitions for a given number n (here, the sum of two dice)
def partitions(n):
    a = [0 for i in range(n + 1)]
    k = 1
    y = n - 1
    while k != 0:
        x = a[k - 1] + 1
        k -= 1
        while 2 * x <= y:
            a[k] = x
            y -= x
            k += 1
        l = k + 1
        while x <= y:
            a[k] = x
            a[l] = y
            yield a[:k + 2]
            x += 1
            y -= 1
        a[k] = x + y
        y = x + y - 1
        yield a[:k + 1]

# Function to clean possible partitions for a given number n such that each element of a partition is unique
def clean_partition(n):
    a = list(partitions(n))
    tile_combos = []
    for tile_options in a:
        if (len(tile_options) == len(set(tile_options)) and max(tile_options) < 11):
            tile_combos.append(tile_options)
    return tile_combos
    
def shut_box(iterations):
    # Write list of tiles. For this game, we assume there are 10 tiles (numbers 1 through 10) to flip
    tiles = [i for i in range (1,11,1)]
    
    
    # Calculate list for sum of two die 
    dice_amounts = [i for i in range (2,13,1)]
    dice_tile_dict = {}

    for amount in dice_amounts:
        dice_tile_dict.update({amount:clean_partition(amount)})
    
    cols = ['roll_amount','flipped_tiles','game_over']
    df = pd.DataFrame(columns=cols)
    
    # Roll two dice
    for i in range(iterations):
        print("Initial Tiles:",tiles)
        dice_1, dice_2 = random.randint(1,6), random.randint(1,6)
        dice_sum = dice_1 + dice_2
        print("Rolled Amount:",dice_sum)
        
        if min(dice_tile_dict[dice_sum],key=len,default='EMPTY') != 'EMPTY':
            tiles_to_flip = min(dice_tile_dict[dice_sum],key=len)
            print("Tile(s) to Flip:",tiles_to_flip)
            
            for tile in tiles_to_flip:
                if tile in tiles:
                    tiles.remove(tile)
                
                for value in dice_tile_dict.values():
                    for l in value:
                        if tile in l:                      
                            value.remove(l)                                           
        
        else:
            df = df.append({'roll_amount':dice_sum,'flipped_tiles':'NaN','game_over':'Lost'},ignore_index=True)
            break
        
        print("Remaining Tiles:",tiles)
        print("***********************************")
        
        df = df.append({'roll_amount':dice_sum,'flipped_tiles':tiles_to_flip},ignore_index=True)
         
                
    
    return df
        
    

shut_box(20)    
```
